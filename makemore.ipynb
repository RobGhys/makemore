## Bigram

What character is likely to follow which character?

```python
for w in words[:1]:
    for ch1, ch2 in zip(w, w[1:]):
        print(ch1, ch2)
```

Why does it work?

zip() pairs up the two iterators and create an iterator of the tuples of the consecutive characters.

```python
for w in words[:1]:
    print(w[1:])
```

```python
w
```

```python
b = {}
for w in words:
    chs = ['<S>'] + list(w) + ['<E>'] # Special start token + (w is a string, so list(w) gives the individual characters) + Special end token
    for ch1, ch2 in zip(chs, chs[1:]):
        bigram = (ch1, ch2)
        b[bigram] = b.get(bigram, 0) + 1 # returns 0 if not in dict, otherwise return frequency
```

```python
b
```

```python
# tuples of key-values
b.items()
```

```python
# Sort by the count of the elements
sorted(b.items(), key = lambda kv: -kv[1])
```

```python
N = torch.zeros((28, 28), dtype=torch.int32)
```

```python
# set of lowercase characters
set(''.join(words))
```

```python
# Sorted list of all words from a to z
chars = sorted(list(set(''.join(words))))

# A mapping from string to int
stoi = {s:i for i,s in enumerate(chars)}
stoi
```

```python
stoi['<S>'] = 26
stoi['<E>'] = 27
```

```python
itos = {i:s for s,i in stoi.items()}
itos
```

```python
for w in words:
    chs = ['<S>'] + list(w) + ['<E>'] # Special start token + (w is a string, so list(w) gives the individual characters) + Special end token
    for ch1, ch2 in zip(chs, chs[1:]):
        ix1 = stoi[ch1]
        ix2 = stoi[ch2]
        N[ix1, ix2] += 1
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.figure(figsize=(16, 16))
plt.imshow(N, cmap='Blues')
for i in range(28):
    for j in range(28):
        # character string (inverse mapping)
        chstr = itos[i] + itos[j]
        # bigram text
        plt.text(j, i, chstr, ha="center", va="bottom", color="gray")
        # number of times the bigram occurs; need .item() to get the int, otherwise it is a tensor
        plt.text(j, i, N[i, j].item(), ha='center', va='top', color='gray')
plt.axis('off')
```

## Refactoring

We make N smaller, i.e. (27,27) vs (28,28

```python
N = torch.zeros((27, 27), dtype=torch.int32)
```

We only use 1 special character, '<.>' instead of 2.

```python
chars = sorted(list(set(''.join(words))))
stoi = {s:i+1 for i,s in enumerate(chars)}
stoi['.'] = 0
itos = {i:s for s,i in stoi.items()}
```

```python
stoi
```

```python
itos
```

```python
for w in words:
    chs = ['.'] + list(w) + ['.']
    for ch1, ch2 in zip(chs, chs[1:]):
        ix1 = stoi[ch1]
        ix2 = stoi[ch2]
        N[ix1, ix2] += 1
```

```python
plt.figure(figsize=(16, 16))
plt.imshow(N, cmap='Oranges')
for i in range(27):
    for j in range(27):
        # character string (inverse mapping)
        chstr = itos[i] + itos[j]
        # bigram text
        plt.text(j, i, chstr, ha="center", va="bottom", color="gray")
        # number of times the bigram occurs; need .item() to get the int, otherwise it is a tensor
        plt.text(j, i, N[i, j].item(), ha='center', va='top', color='gray')
plt.axis('off')
```

### Use a multinomial distribution with a generator (so that results are deterministic)

```python
g = torch.Generator().manual_seed(2147483647)
p = torch.rand(3, generator=g)
p = p/p.sum()
```

```python
p
```

```python
assert(p.sum() == 1.0)
```

```python
torch.multinomial(p, num_samples=100, replacement=True, generator=g)
```

```python
g = torch.Generator().manual_seed(2147483647)
# index (need to use .item() to get the int)
ix = torch.multinomial(p, num_samples=1, replacement=True, generator=g).item()
ix
```

```python
# Which character we are sampling
itos[ix]
```

```python
g = torch.Generator().manual_seed(2147483647)

for i in range(10):
    out = []
    # start at index 0
    ix = 0
    while True:
        p = N[ix].float()
        p = p/p.sum() # normalize the value to get a probability
        # Next index, sample from distribution
        ix = torch.multinomial(p, num_samples=1, replacement=True, generator=g).item()
        out.append(itos[ix])
        if ix == 0:
            break
    print(''.join(out))
```
